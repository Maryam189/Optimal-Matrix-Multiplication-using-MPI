```python
import matplotlib.pyplot as plt

# Results data
processes = [2, 4, 6, 8]
one_machine_times = [0.014632, 0.003362, 0.014425, 0.035315]
two_machines_times = [0.000763, 1.163507, 0.064302, 0.103783]

# Plotting the results
plt.plot(processes, one_machine_times, marker='o', label='One Machine')
plt.plot(processes, two_machines_times, marker='o', label='Two Machines')

# Adding labels and title
plt.xlabel('Number of Processes')
plt.ylabel('Execution Time (seconds)')
plt.title('Blocking Matrix Multiplication Execution Time')

# Adding legend
plt.legend()

# Display the plot
plt.show()

```
